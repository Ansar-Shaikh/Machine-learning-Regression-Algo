```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("knn_regression_dataset.csv")
df
```

```python
df.isnull().sum()
```

```python
x=df.drop(columns="Salary")
y=df["Salary"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=15)
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
for i in range(1,30):
    knc1=KNeighborsRegressor(n_neighbors=i)
    knc1.fit(x_train,y_train)
    print(i,knc1.score(x_train,y_train)*100,knc1.score(x_test,y_test)*100)
```

```python
knc=KNeighborsRegressor(n_neighbors=30)
knc.fit(x_train,y_train)
knc.score(x_train,y_train)*100
```
