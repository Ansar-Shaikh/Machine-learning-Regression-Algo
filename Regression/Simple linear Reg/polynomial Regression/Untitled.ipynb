```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_csv("Position_Salaries.csv")
df.head()
df.drop(columns="Position",axis=1,inplace=True)
df
```

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1:]
x,y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

# Training the Linear Regression model on the whole dataset

```python
from sklearn.linear_model import LinearRegression
lr1=LinearRegression()
lr1.fit(x,y)
```

# Training the Polynomial Regression model on the whole dataset

```python
from sklearn.preprocessing import PolynomialFeatures
pf=PolynomialFeatures(degree=4)
x_poly=pf.fit_transform(x)
lr2=LinearRegression()
lr2.fit(x_poly,y)
```

```python
plt.scatter(df["Level"],df["Salary"])
plt.show()
```

```python
plt.scatter(x, y, color = 'red')
plt.plot(x, lr1.predict(x), color = 'blue')
plt.title('Truth or Bluff (Linear Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

# it not show linearty between datapoints  use polynomial function

```python

plt.scatter(x,y, color = 'red')
y_poly=pf.fit_transform(x)
plt.plot(x, lr2.predict(y_poly),y_poly, color = 'blue')
plt.title('Truth or Bluff (Polynomial Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

```python
lr1.score(x_test,y_test)* 100 # to accuracy of  model 
```
