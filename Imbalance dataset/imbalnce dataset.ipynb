# what is imbalncing model get trained on the class 0 having repeatedly occured majority is class 0 model get baisded on class 0 therefore model get trained on class 0 when unseen data passedd to model it predict by default to class 0 this is called is imbalacing data

```python
x=df.iloc[:,:-1]
y=df["purchased"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
print("accuracy of model :",lr.score(x_test,y_test)*100)
```

```python
 # to check imbalancing
df["purchased"].value_counts()
```
