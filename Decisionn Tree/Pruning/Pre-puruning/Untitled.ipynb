# dt=DecisionTreeClassifier(max-depth=any depth) thats is prepurning 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("pruned_social_media_dataset.csv")
df
```

```python
df.isnull().sum()
```

```python
x=df.iloc[:,:-1]
y=df["purchased"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt=DecisionTreeClassifier(criterion="entropy",max_depth=3) #prepurining 
```

```python
dt.fit(x_train,y_train)
```


# check first model get overfitted or not 

# pruning is done when model get over fitted 

```python
print("Check accuracy between testing dataset and training data if high diff then it is overfitted::",dt.score(x_train,y_train)*100)
```

```python
print("Check accuracy between testing dataset and training data if high diff then it is overfitted::",dt.score(x_test,y_test)*100)
```

# on training  dataset model accuracy is 100 percent  but on testing dataset model will not capeable to much accurate as trained perform that is model get overfitted we need to pruning it 

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=dt)
plt.show()
```

```python
from sklearn.tree import plot_tree
```

```python
plot_tree(dt)
plt.show()
```
