# Mode fill in place of NaN values ,It works only and only one column


```python
df["species"].mode()
```

```python
df["species"].mode()[0]
```

```python
df["species"].fillna(df["species"].mode()[0])
```

# But mode work at a time one col.therefore to work on whole object datatype in dataset we use loop concepts


```python
index=df.select_dtypes(include="object").columns
for i in index:
    df[i].fillna(df[i].mode()[0])
```
